```python
#Request via Web Driver
options = Options()
options.headless = True
options.add_argument("--window-size=1920,1200")
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get("https://www.gianttiger.com/search?q=Bacon+Sausages+%26+Hot+Dogs&L3_category_en=Bacon+Sausages+%26+Hot+Dogs")
sleep(15)
r = driver.page_source
soup = BeautifulSoup(r, "html.parser")
```

```python
import pandas as pd

# create an empty DataFrame
df = pd.DataFrame(columns=['URL'])

for a in soup.find_all("a", {"class": "product-tile__link inline-link-small"}):
    # concatenate the base URL with the href value
    url = "https://www.gianttiger.com/" + a['href']
    # create a new DataFrame with the URL value
    new_row = pd.DataFrame({'URL': [url]})
    # concatenate the new DataFrame with the main DataFrame
    df = pd.concat([df, new_row], ignore_index=True)

# print the resulting DataFrame
df
```

```python
# list of URLs
urls = df.values.tolist()       
        
driver = webdriver.Chrome(ChromeDriverManager().install())
data_list = []

for url in urls:
    # send a GET request to the URL and get the HTML content
    driver.get(url)
    sleep(6)
    r = driver.page_source
    soup = BeautifulSoup(r, "html.parser")


    # Retrieve product information
    name = soup.find('h1',class_='product__title').text
    Specification = soup.select('div#collapse-product-specifications ul li')
    price = soup.find('span',class_='price__value').text
    categoryList = soup.find('ol', class_='breadcrumbs__list')
    category = categoryList.find_all('li')[-1].text

    #images
    images = []
    for x in soup.findAll('img', attrs={'srcset': True}):
        srcset = x['srcset']
        image = re.match(r'^([^,\s]+).*$', srcset).group(1)
        images.append(image)

    # create a list to store the text content of each `li` element
    specs_list = []
    lis = soup.select('div#collapse-product-details ul li')
    for li in lis:
        specs_list.append(li.text.strip())

    # create dictionary with the extracted information
    data = {'Name': name, 'Price': price, 'Category': category,'Product details': specs_list,'image': images}

    # append dictionary to list
    data_list.append(data)

# create DataFrame from list of dictionaries
df = pd.DataFrame(data_list)
```
